# Importing libraries

```python
import shap
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.metrics import classification_report, confusion_matrix
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization
from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping

import keras_tuner as kt

import seaborn as sns
import pandas as pd
from imblearn.over_sampling import SMOTE
```

# Loading data

```python
# 1. Cargar datos
liver_cancer_dataset = pd.read_csv('../0. Data/0. Original/liver_data_mar22_d032222.csv')
etiquetas = liver_cancer_dataset.liver_cancer

df_imputed = liver_cancer_dataset.copy()

# Eliminar columnas no necesarias
datos = df_imputed.drop(columns=['liver_cancer', 'liver_topography', 'plco_id', 'build'])
```

```python
feature_names = datos.columns.tolist()
```

# Imputation

```python
# 2. Imputación de valores faltantes con la media
imputer = SimpleImputer(strategy="mean")
datos_imputed = pd.DataFrame(imputer.fit_transform(datos), columns=datos.columns)
```

# Data balancing

```python
# 3. División en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(datos_imputed, etiquetas, test_size=0.2, random_state=42)

# 4. Balanceo de clases con SMOTE
smote = SMOTE(random_state=42)
X_train_res, y_train_res = smote.fit_resample(X_train, y_train)
```

# Normalization

```python
# 5. Normalización
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train_res)
X_test = scaler.transform(X_test)
```

# Creating the model

```python
# 6. Definir el modelo
def build_model(hp):
    model = Sequential()
    
    # Capa densa inicial
    model.add(Dense(
        units=hp.Int('units_layer1', min_value=32, max_value=128, step=16),  # Busca entre 32 y 128 unidades
        activation='relu',
        input_dim=X_train.shape[1]
    ))
    model.add(BatchNormalization())
    model.add(Dropout(hp.Float('dropout_layer1', min_value=0.2, max_value=0.5, step=0.1)))  # Dropout entre 0.2 y 0.5

    # Capa densa intermedia
    model.add(Dense(
        units=hp.Int('units_layer2', min_value=16, max_value=64, step=16),  # Busca entre 16 y 64 unidades
        activation='relu'
    ))
    model.add(BatchNormalization())
    model.add(Dropout(hp.Float('dropout_layer2', min_value=0.2, max_value=0.5, step=0.1)))  # Dropout entre 0.2 y 0.5

    # Capa de salida
    model.add(Dense(1, activation='sigmoid'))

    # Compilación del modelo
    model.compile(
        optimizer=Adam(learning_rate=hp.Float('learning_rate', min_value=1e-4, max_value=1e-2, sampling='log')),  # LR entre 1e-4 y 1e-2
        loss='binary_crossentropy',
        metrics=['Recall', 'Precision', 'accuracy']  # Usamos Recall y Accuracy como métrica
    )
    
    return model
```

# Hyperparameters tuning

```python
# 7. Búsqueda de hiperparámetros con Keras Tuner
tuner = kt.Hyperband(
    build_model,
    objective=kt.Objective("val_recall", direction="max"),  # Especificamos que se debe maximizar
    max_epochs=20,
    factor=3, # Controla la cantidad de pruebas en cada ciclo de Hyperband
    project_name="full_dataset_tuning_v1"  # Nuevo nombre de proyecto
)

# 8. Ejecutar la búsqueda de hiperparámetros
tuner.search(X_train, y_train_res, epochs=20, validation_split=0.2, verbose=1)

# 8. Obtener los mejores hiperparámetros encontrados
best_hps = tuner.get_best_hyperparameters(num_trials=1)[0]

# 9. Construir el modelo con los mejores hiperparámetros
model = build_model(best_hps)
```

# Model Training and Evaluation

```python
# 7. Entrenar el modelo con callbacks
callbacks = [
    keras.callbacks.EarlyStopping(
        monitor='val_loss', patience=10, restore_best_weights=True, verbose=1
    ),
    keras.callbacks.ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=5)
]

class_weights = {0: 1, 1: 6}  # Dar más peso a la clase minoritaria
history = model.fit(X_train, y_train_res, validation_split=0.2, epochs=20, class_weight=class_weights, callbacks=callbacks)

# 8. Evaluación del modelo
y_pred = (model.predict(X_test) > 0.4).astype(int)
print(classification_report(y_test, y_pred))
```

# Plotting metrics

```python
# 9. Matriz de confusión
conf_matrix = confusion_matrix(y_test, y_pred)
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues')
plt.xlabel("Predicho")
plt.ylabel("Real")
plt.title("Matriz de Confusión")
plt.show()

from sklearn.metrics import roc_auc_score
auc = roc_auc_score(y_test, model.predict(X_test))
print("AUC:", auc)
```

```python
def plot_training_curves(history):
    epochs = np.arange(len(history.history['loss']))
    
    # Encontrar la mejor época (mínima val_loss)
    best_epoch = np.argmin(history.history['val_loss'])

    # Crear una sola figura con 3 subgráficos en una fila
    plt.figure(figsize=(12, 18))

    # Figura de pérdidas (Loss)
    plt.subplot(3, 1, 1)
    plt.plot(epochs, history.history['loss'], label='Entrenamiento')
    plt.plot(epochs, history.history['val_loss'], label='Validación')
    plt.axvline(best_epoch, linestyle='--', color='r', label=f'Mejor época: {best_epoch}')
    plt.xlabel('Épocas')
    plt.ylabel('Pérdida')
    plt.title('Curva de Pérdida')
    plt.legend()
    plt.grid(True)


    # Figura de Recall (Recall) 
    plt.subplot(3, 1, 2)
    plt.plot(history.history['recall'], label='Train Recall')
    plt.plot(history.history['val_recall'], label='Val Recall')
    plt.axvline(best_epoch, linestyle='--', color='r', label=f'Mejor época: {best_epoch}')
    plt.xlabel("Epochs")
    plt.ylabel("Recall")
    plt.title("Curva de Recall")
    plt.legend()
    plt.grid(True)

    # Figura de precisión (Precision)
    plt.subplot(3, 1, 3)
    plt.plot(epochs, history.history['precision'], label='Entrenamiento')
    plt.plot(epochs, history.history['val_precision'], label='Validación')
    plt.axvline(best_epoch, linestyle='--', color='r', label=f'Mejor época: {best_epoch}')
    plt.xlabel('Épocas')
    plt.ylabel('Precisión')
    plt.title('Curva de Precisión')
    plt.legend()
    plt.grid(True)

    # Ajustar los espacios entre gráficos para que tengan el mismo tamaño
    plt.subplots_adjust(hspace=0.4)  # Ajustamos espacio vertical entre los gráficos

    plt.show()

# Llamar a la función después del entrenamiento
plot_training_curves(history)
```
