# Importing libraries

```python
import pandas as pd
import numpy as np

## Plot
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt

## Selection Models
from sklearn.model_selection import StratifiedKFold

## Models
from sklearn.ensemble import RandomForestClassifier

## Metrics
from sklearn.metrics import make_scorer, accuracy_score, balanced_accuracy_score, precision_score, recall_score, f1_score
from sklearn.metrics import confusion_matrix

## Dataset balancing
from imblearn.over_sampling import SMOTE, RandomOverSampler
from imblearn.under_sampling import NearMiss
```

# Loading Data

```python
## Imputed dataset
imputed_dataset = pd.read_csv('../../0. Data/2. Imputed/dataset_imputed_mean_median.csv')

# Eliminar la variable target
imputed_dataset_copy = imputed_dataset.copy()

imputed_dataset_no_target = imputed_dataset_copy.drop(columns=['liver_cancer'])
imputed_dataset_no_target_inputs = imputed_dataset_no_target

pd.set_option('display.max_columns', None)
```

```python
target = imputed_dataset.liver_cancer
target.info()
```

## Dataset balancing

```python
"""
    Función genérica para comparar métodos (balanceo, algoritmos, etc.) y seleccionar el mejor, incluyendo matriz de confusión.

    Parámetros:
    -----------
    methods : dict
        Diccionario con los nombres de los métodos y sus funciones (balanceo, algoritmos, etc.).
    X : DataFrame o array
        Conjunto de características.
    y : array
        Variable objetivo.
    cv_splits : int
        Número de pliegues para validación cruzada (default: 5).
    task_name : str
        Nombre de la tarea evaluada (e.g., "Balancing Method", "Model").
    include_conf_matrix : bool
        Si es True, incluye la matriz de confusión promedio en los resultados (default: True).

    Retorna:
    --------
    tuple:
        - DataFrame con los resultados de todas las evaluaciones.
"""

def evaluate_models_methods(methods, X, y, cv_splits=5, task_name="Method", metrics=None, include_conf_matrix=True):
    if metrics is None:
        metrics = {
            "Accuracy": make_scorer(accuracy_score),
            "Balanced Accuracy": make_scorer(balanced_accuracy_score)
        }

    # Comparar resultados
    results = []
    cv = StratifiedKFold(n_splits=cv_splits, shuffle=True, random_state=42)

    for method_name, method_func in methods.items():

        # Validación cruzada con el modelo
        print(f"Evaluando: {task_name} = {method_name}")

        # Aplicar el método a los datos
        X_transformed, y_transformed = method_func(X, y)
        X_transformed = np.array(X_transformed)
        y_transformed = np.array(y_transformed)

        method_results = {task_name: method_name}
        
        y_pred_combined = []  # Para almacenar predicciones combinadas en todos los pliegues
        y_test_combined = []  # Para almacenar valores reales combinados en todos los pliegues

        for metric_name, scorer in metrics.items():
            # Validación cruzada manual
            scores = []
            for train_idx, test_idx in cv.split(X_transformed, y_transformed):
                X_train, X_test = X_transformed[train_idx], X_transformed[test_idx]
                y_train, y_test = y_transformed[train_idx], y_transformed[test_idx]
                
                # Modelo base
                model = RandomForestClassifier(random_state=42)
                model.fit(X_train, y_train)
                y_pred = model.predict(X_test)

                # Almacenar valores reales y predicciones para la matriz de confusión
                y_test_combined.extend(y_test)
                y_pred_combined.extend(y_pred)
                
                # Calcular métrica
                score = scorer._score_func(y_test, y_pred)
                scores.append(score)
                    
            method_results[f"{metric_name}"] = ": %0.5f (+/- %0.5f)" % (np.mean(scores), np.std(scores) * 2)

        # Graficar la matriz de confusión usando la función proporcionada
        if include_conf_matrix:
            title = f"Matriz de Confusión - {method_name}"
            plot_confusion_matrix(y_test_combined, y_pred_combined, title)
        
        results.append(method_results)
    
    # Convertir resultados a DataFrame
    results_df = pd.DataFrame(results)
    return results_df

# Función para graficar matriz de confusión
def plot_confusion_matrix(y_true, y_pred, title):
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(6, 4))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=["No Liver Cancer", "Liver Cancer"], yticklabels=["No Liver Cancer", "Liver Cancer"])
    plt.title(title)
    plt.xlabel("Predicción")
    plt.ylabel("Valor Real")
    plt.show()
```

```python
# Separar características (X) y variable objetivo (y)
X = imputed_dataset_no_target_inputs
y = target

# Filtrar solo columnas numéricas
X = X.select_dtypes(include=['number'])
balancing_methods = {
    "Original": lambda X, y: (X, y),
    "SMOTE": lambda X, y: SMOTE(random_state=42).fit_resample(X, y),
    "Oversampling": lambda X, y: RandomOverSampler(random_state=42).fit_resample(X, y),
    "Undersampling": lambda X, y: NearMiss(version=3).fit_resample(X, y)
}

metrics = {
        "Accuracy": make_scorer(accuracy_score),
        "Balanced Accuracy": make_scorer(balanced_accuracy_score),
        "Precision (Class 1)": make_scorer(precision_score, pos_label=1),
        "Recall (Class 1)": make_scorer(recall_score, pos_label=1),
        "F1-Score (Class 1)": make_scorer(f1_score, pos_label=1)
    }

# Evaluar métodos de balanceo
balance_results = evaluate_models_methods(balancing_methods, X, y, cv_splits=5, task_name="Method", metrics=metrics, include_conf_matrix=True)

# Mostrar resultados
print(balance_results)
```

```python
# Seleccionar el mejor método basado en F1-Score (Clase 1)
best_method = balance_results.sort_values(by="F1-Score (Class 1)", ascending=False).iloc[0]

print("\n=== Mejor Método de Balanceo ===")
print(best_method)
```
